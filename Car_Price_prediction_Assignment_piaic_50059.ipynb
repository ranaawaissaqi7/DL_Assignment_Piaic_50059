```python
import tensorflow as tf
from keras import models
from keras import layers
def build_model(act):
 model = models.Sequential()
 model.add(layers.Dense(128, activation=act,input_shape=(199,)))
 model.add(layers.Dense(64, activation=act))
 model.add(layers.Dense(32, activation=act))
 model.add(layers.Dense(1))
 model.compile(optimizer='adam', loss='mse', metrics=['mae'])
 return model
```

```python
import numpy as np
k = 4
num_val_samples = len(x_train_mmc) // k
num_epochs = 100
all_scores = []
for i in range(k):
 print('processing fold #', i)
 val_data = x_train_mmc[i * num_val_samples: (i + 1) * num_val_samples]
 val_targets = y_train_mmc[i * num_val_samples: (i + 1) * num_val_samples]
 partial_train_data = np.concatenate([x_train_mmc[:i * num_val_samples],x_train_mmc[(i + 1) * num_val_samples:]],axis=0)
 partial_train_targets = np.concatenate([y_train_mmc[:i * num_val_samples],y_train_mmc[(i + 1) * num_val_samples:]],axis=0)
 model = build_model('tanh')
 model.fit(partial_train_data, partial_train_targets,epochs=num_epochs, batch_size=1, verbose=0)
 val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0)
 all_scores.append(val_mae)
```
